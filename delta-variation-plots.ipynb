```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Garuda"
plt.rcParams["font.size"] = 11
```

```python
deltas = ["10", "1", "0p1"]

data = {}

for delta in deltas:
    data[delta] = np.genfromtxt(f"data/delta/rho0p5/{delta}.txt")
```

```python
L = 300
t_f = 3000
bins = 300

time_steps = np.arange(t_f)

# Dictionaries to store histograms and edges
histograms = {}
xedges = {}
yedges = {}

for delta in deltas:
    data_flattened = data[delta].flatten()
    hist, xedge, yedge = np.histogram2d(data_flattened, np.repeat(time_steps, data[delta].shape[1]), bins=[bins, t_f], range=[[0, L], [0, t_f]])
    histograms[delta] = hist
    xedges[delta] = xedge
    yedges[delta] = yedge
```

```python
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(12, 4.5))
fig.suptitle("Distribución de caminantes en el tiempo variando el rango de interacción ($\\Delta$) para la densidad $\\rho=0.05$.\nSeed 2906", fontweight="bold", fontsize=15)
fig.supxlabel("Posición")
fig.supylabel("Paso de tiempo")

for a in ax.flat:
    a.tick_params(axis='y', which='major', rotation=60)

titles = ["$\\Delta=10$", "$\\Delta=1$", "$\\Delta=0.1$"]

for a, title, eta in zip(ax.flat, titles, deltas):
    a.set_title(title)
    im = a.imshow(histograms[eta].T, origin='lower', aspect='auto', extent=[0, L, 0, t_f], cmap='RdGy_r')
    # im = a.imshow(histogram.T, origin='lower', aspect='auto', extent=[0, L, 0, t_f], cmap='pink')

    cbar = fig.colorbar(im, ax=a)
    cbar.ax.tick_params(rotation=60)

fig.tight_layout()
fig.savefig("figs/delta-rho0p5.pdf")
```
