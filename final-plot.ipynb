```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Garuda"
plt.rcParams["font.size"] = 11
```

```python
data = np.genfromtxt("data/final.txt")
data6 = np.genfromtxt("data/final6.txt")
```

```python
L = 300
t_f = 3000
bins = 300

time_steps = np.arange(t_f)

data_flattened = data.flatten()
hist, xedge, yedge = np.histogram2d(data_flattened, np.repeat(time_steps, data.shape[1]), bins=[bins, t_f], range=[[0, L], [0, t_f]])

data6_flattened = data6.flatten()
hist6, xedge6, yedge6 = np.histogram2d(data6_flattened, np.repeat(time_steps, data6.shape[1]), bins=[bins, t_f], range=[[0, L], [0, t_f]])
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.set_title("Distribucin de caminantes en el tiempo. Seed 2906", fontweight="bold", fontsize=15)
ax.set_xlabel("Posicin")
ax.set_ylabel("Paso de tiempo")

ax.tick_params(axis='y', which='major', rotation=60)

im = ax.imshow(hist.T, origin='lower', aspect='auto', extent=[0, L, 0, t_f], cmap='RdGy_r')
# im = a.imshow(histogram.T, origin='lower', aspect='auto', extent=[0, L, 0, t_f], cmap='pink')

cbar = fig.colorbar(im, ax=ax)
cbar.ax.tick_params(rotation=60)

fig.tight_layout()
fig.savefig("figs/final2.pdf")
```
